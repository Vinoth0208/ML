# Linear Regression

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Vinoth\Downloads\homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area (sqft)')
plt.ylabel('Price ($)')
plt.scatter(df.area,df.price, color='red', marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[6000]])
```

# Prediction

```python
dff=pd.read_csv(r"C:\Users\Vinoth\Downloads\areas.csv")
dff
```

```python
y_pred=reg.predict(dff[['area']])
```

```python
dff['y_pred']=y_pred
dff
```

```python
dff.to_csv(r"Houserate.csv")
```
